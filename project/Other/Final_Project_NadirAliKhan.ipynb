## Using Pandas Library to Read, Clean, Filter the Data

```python
import pandas as pd 

data = pd.read_csv("Travel.csv")
df = pd.DataFrame(data)




# There is typo error in Gender Column "Female" > "Fe Male" 
print(df['Gender'].unique())

# Replacing the "Fe Male" with "Female"
df['Gender'] = df['Gender'].replace('Fe Male', 'Female')


# There are some columns with missing values
print(df.isnull().sum())

# getting first five rows of the data 
df.head()
```

# Filling Missing Values with Pandas

```python

# Filling numerical columns with median
df['Age'] = df['Age'].fillna(df['Age'].median())
df['DurationOfPitch'] = df['DurationOfPitch'].fillna(df['DurationOfPitch'].median())
df['NumberOfFollowups'] = df['NumberOfFollowups'].fillna(df['NumberOfFollowups'].median())
df['NumberOfTrips'] = df['NumberOfTrips'].fillna(df['NumberOfTrips'].median())
df['NumberOfChildrenVisiting'] = df['NumberOfChildrenVisiting'].fillna(df['NumberOfChildrenVisiting'].median())
df['MonthlyIncome'] = df['MonthlyIncome'].fillna(df['MonthlyIncome'].median())

# Filling categorical columns with mode
df['TypeofContact'] = df['TypeofContact'].fillna(df['TypeofContact'].mode()[0])
df['PreferredPropertyStar'] = df['PreferredPropertyStar'].fillna(df['PreferredPropertyStar'].mode()[0])

# Checking filled missing values
print(df.isnull().sum())
```

# Project Task 2:
## Customer Conversion Rate by Contact Type

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load your dataset
# Assuming df is your DataFrame

# Step 1: Group by 'TypeofContact' and calculate conversion rates
conversion_data = df.groupby('TypeofContact')['ProdTaken'].agg(['sum', 'count'])
conversion_data['ConversionRate'] = (conversion_data['sum'] / conversion_data['count']) * 100

# Reset index to make it easier to plot
conversion_data.reset_index(inplace=True)

# Step 2: Visualization using Matplotlib
plt.figure(figsize=(8, 4))
plt.bar(conversion_data['TypeofContact'], conversion_data['ConversionRate'], color=['#1f77b4', '#ff7f0e'])
plt.title('Customer Conversion Rate by Contact Type')
plt.xlabel('Type of Contact')
plt.ylabel('Conversion Rate (%)')
plt.ylim(0, 100)  # Set y-axis from 0 to 100 for percentage
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# Optional: Display the conversion data
print(conversion_data)
```

# Project Task 3:
## Travel Frequency by Age Group

```python
# Creating age groups
bins = [18, 25, 35, 45, 60, 100]
labels = ['18-25', '26-35', '36-45', '46-60', '60+']
df['AgeGroup'] = pd.cut(df['Age'], bins=bins, labels=labels, right=False)


# Group by AgeGroup and calculate the mean of NumberOfTrips
travel_frequency = df.groupby('AgeGroup')['NumberOfTrips'].mean().reset_index()
```

```python
import matplotlib.pyplot as plt

# Plotting the travel frequency by age group
plt.figure(figsize=(8, 5))
plt.bar(travel_frequency['AgeGroup'], travel_frequency['NumberOfTrips'], color='teal')
plt.xlabel('Age Group')
plt.ylabel('Average Number of Trips')
plt.title('Travel Frequency by Age Group')
plt.show()
```

# Project Task 4:
## Satisfaction Score Analysis

```python
import matplotlib.pyplot as plt

# Plotting the distribution of PitchSatisfactionScore
plt.figure(figsize=(8, 5))
plt.hist(df['PitchSatisfactionScore'], bins=10, color='skyblue', edgecolor='black')
plt.xlabel('Pitch Satisfaction Score')
plt.ylabel('Frequency')
plt.title('Distribution of Pitch Satisfaction Scores')
plt.show()
```

```python
# Grouping by Type of Contact and calculating the average satisfaction score
contact_satisfaction = df.groupby('TypeofContact')['PitchSatisfactionScore'].mean().reset_index()

# Plotting the results
plt.figure(figsize=(8, 5))
plt.bar(contact_satisfaction['TypeofContact'], contact_satisfaction['PitchSatisfactionScore'], color='orange')
plt.xlabel('Type of Contact')
plt.ylabel('Average Pitch Satisfaction Score')
plt.title('Average Satisfaction Score by Contact Type')
plt.show()
```

```python
# Grouping by Product Pitched and calculating the average satisfaction score
product_satisfaction = df.groupby('ProductPitched')['PitchSatisfactionScore'].mean().reset_index()

# Plotting the results
plt.figure(figsize=(12, 6))
plt.bar(product_satisfaction['ProductPitched'], product_satisfaction['PitchSatisfactionScore'], color='green')
plt.xlabel('Product Pitched')
plt.ylabel('Average Pitch Satisfaction Score')
plt.title('Average Satisfaction Score by Product Pitched')
plt.xticks(rotation=45)
plt.show()
```

```python
import numpy as np

# Scatter plot to observe the relationship between Duration of Pitch and Satisfaction Score
plt.figure(figsize=(8, 6))
plt.scatter(df['DurationOfPitch'], df['PitchSatisfactionScore'], alpha=0.6, color='purple')
plt.xlabel('Duration of Pitch (minutes)')
plt.ylabel('Pitch Satisfaction Score')
plt.title('Duration of Pitch vs. Satisfaction Score')
plt.grid(True)
plt.show()
```

```python
# Grouping by Age Group and calculating the average satisfaction score
age_satisfaction = df.groupby('AgeGroup')['PitchSatisfactionScore'].mean().reset_index()

# Plotting the results
plt.figure(figsize=(8, 6))
plt.bar(age_satisfaction['AgeGroup'], age_satisfaction['PitchSatisfactionScore'], color='dodgerblue')
plt.xlabel('Age Group')
plt.ylabel('Average Pitch Satisfaction Score')
plt.title('Average Satisfaction Score by Age Group')
plt.show()
```
