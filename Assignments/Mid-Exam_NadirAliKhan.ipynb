```python
import numpy as np
# three arrays
np.random.seed(42)
year_19 = np.random.randint(5, 50, size=12)
year_20 = np.random.randint(5, 50, size=12)
year_21 = np.random.randint(5, 50, size=12)

```

```python
import matplotlib.pyplot as plt
# a
plt.plot(year_19)
plt.plot(year_20)
plt.plot(year_21)
plt.title("Temp: for year 19, 20, 21")
plt.ylabel("Temp:")
plt.xlabel("Days")
plt.legend(["19", "20", "21"])
plt.show()
```

```python
# b
# avgs
avg_19 = year_19.mean()
avg_20 = year_20.mean()
avg_21 = year_21.mean() 

plt.bar(["2019", "2020", "2021"],[avg_19, avg_20, avg_21] )
plt.show()
```

```python
# c

plt.hist(year_21, bins=5, width=6)
plt.show()
```
