```python
# Write a Python program that calculates the grade for a student based on the following criteria:

# Greater than or equal to 90: "A"
# Between 80 and 89: "B"
# Between 70 and 79: "C"
# Between 60 and 69: "D"
# Below 60: "F"

student_grade = int(input("Enter student score (0-100): "))

if student_grade > 100 or student_grade < 0:
    print("Invalid score. Please enter a score between 0 and 100.")
else:
    if student_grade >= 90:
        grade = "A"
    elif student_grade >= 80:
        grade = "B"
    elif student_grade >= 70:
        grade = "C"
    elif student_grade >= 60:
        grade = "D"
    else:
        grade = "F"

    print(f"Grade: {grade}")


```
