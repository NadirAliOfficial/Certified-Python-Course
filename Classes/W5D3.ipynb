```python
import pandas as pd
import numpy as np

# set col as index

data = pd.read_csv("data/gdp.csv", index_col='country')
df = pd.DataFrame(data)
df.head()
```

```python
# plotting the data for norway

import matplotlib.pylab as plt
plt.bar(df.columns, df.loc['Norway'])
plt.xticks(rotation=65)
plt.yticks(rotation=65)
plt.title('Norway')

```
