```python
import matplotlib.pyplot as plt
import numpy  as np

# Create a 10x10 array of random integers between 0 and 100

course = input("Slect Course: \n1.Python\n2.Java.\n3.Data Science")

if course =='1':
    print("Python Course Selected")
elif course =='2':
    print("Java Course Selected")
elif course =='3':
    print("Data Science Course Selected")


```

```python
y = np.random.randint(1,6, size=6)
x = np.array([1,2,3,4,5,6])
print("Array: ", x,y)
# plt.plot(kind='line', figsize=(4,2),arr)
plt.plot(x,y, marker='o', color='k', ls='--', ms=10, mec='b', lw=2, mfc = 'b')
```

```python
days = [1,2,3,4,5,6,7]
temp = [34, 36, 32, 30, 29, 39, 40]
hmd = [40, 45, 43, 49, 46, 50, 51]
wind = np.random.randint(30,40, size=7)
plt.plot(days, temp, marker='o', mec='b', mfc='b', label='temp')
plt.figure()
plt.plot(days, hmd,  marker='o', mec='r', mfc='r', label='hmd')
plt.figure()
plt.plot(days, wind,  marker='o', mec='k', mfc='k', label='wind')
plt.xlabel("Days", fontsize=15)
plt.ylabel("Temperature", fontsize=15)
plt.legend()
plt.title("Temperature Graph", fontsize=20)
# for adding bg to graph
plt.gcf().set_facecolor('black') 
```

```python
data = np.random.randint(50, 250, size=5)
x = np.arange(0, len(data))
plt.bar(x, data, color='k')
plt.xlabel("Product Names", fontsize=15)
plt.ylabel("Sales", fontsize=15)
plt.title("Sales Analysis", fontsize=20)
plt.show()
```

```python
x = np.arange(1,6)
y = np.random.randint(1,10, size=5)
plt.scatter(x,y, marker='o', c='r', s=50)
plt.grid()
```

```python
data_set = np.random.randint(1,100, size=1000)
plt.hist(data_set, bins=20, alpha=0.3)
plt.title('Histogram', fontsize=27)
plt.xlabel('Numbers')
plt.ylabel('Numbers')
```

```python
prc = np.random.randint(10,30, size=5)
fru = ["apples", "bananas", "oranges", "grapes", "pears"]
plt.pie(prc, labels=fru, autopct='%1.1f%%')
```
