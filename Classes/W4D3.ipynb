```python
from datetime import datetime

# Function to get the astrological sign
def get_zodiac_sign(day, month):
    if (month == 12 and day >= 22) or (month == 1 and day <= 19):
        return 'Capricorn', 'Responsible, disciplined, self-control, good managers.'
    elif (month == 1 and day >= 20) or (month == 2 and day <= 18):
        return 'Aquarius', 'Progressive, original, independent, humanitarian.'
    elif (month == 2 and day >= 19) or (month == 3 and day <= 20):
        return 'Pisces', 'Compassionate, artistic, intuitive, gentle, wise.'
    elif (month == 3 and day >= 21) or (month == 4 and day <= 19):
        return 'Aries', 'Courageous, determined, confident, enthusiastic.'
    elif (month == 4 and day >= 20) or (month == 5 and day <= 20):
        return 'Taurus', 'Reliable, patient, practical, devoted, responsible.'
    elif (month == 5 and day >= 21) or (month == 6 and day <= 20):
        return 'Gemini', 'Gentle, affectionate, curious, adaptable, quick learner.'
    elif (month == 6 and day >= 21) or (month == 7 and day <= 22):
        return 'Cancer', 'Tenacious, highly imaginative, loyal, emotional, sympathetic.'
    elif (month == 7 and day >= 23) or (month == 8 and day <= 22):
        return 'Leo', 'Creative, passionate, generous, warm-hearted, cheerful.'
    elif (month == 8 and day >= 23) or (month == 9 and day <= 22):
        return 'Virgo', 'Loyal, analytical, kind, hardworking, practical.'
    elif (month == 9 and day >= 23) or (month == 10 and day <= 22):
        return 'Libra', 'Cooperative, diplomatic, gracious, fair-minded, social.'
    elif (month == 10 and day >= 23) or (month == 11 and day <= 21):
        return 'Scorpio', 'Resourceful, brave, passionate, stubborn, a true friend.'
    elif (month == 11 and day >= 22) or (month == 12 and day <= 21):
        return 'Sagittarius', 'Generous, idealistic, great sense of humor.'

# Main program to input DOB and print the sign and traits
def main():
    try:
        dob_str = input("Enter your date of birth (DD-MM-YYYY): ")
        dob = datetime.strptime(dob_str, "%d-%m-%Y")
        
        # Get the day and month from DOB
        day = dob.day
        month = dob.month
        
        # Get the zodiac sign and traits
        zodiac_sign, traits = get_zodiac_sign(day, month)
        
        # Print the zodiac sign and traits
        print(f"\nYour zodiac sign is: {zodiac_sign}")
        print(f"Traits: {traits}")
    
    except ValueError:
        print("Invalid date format. Please use DD-MM-YYYY.")

if __name__ == "__main__":
    main()
```
